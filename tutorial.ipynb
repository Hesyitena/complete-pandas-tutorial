```python
# Fungsi untuk mengkategorikan atlet berdasarkan kombinasi tinggi dan berat badan
def categorize_athlete(row):
    # Jika tinggi < 175cm DAN berat < 70kg
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    # Jika tinggi < 185cm ATAU berat <= 80kg
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'Middleweight'
    # Jika tidak memenuhi keduanya (badan besar dan berat)
    else:
        return 'Heavyweight'

# Menerapkan fungsi ke setiap baris (axis=1) dan simpan di kolom baru 'Category'
bios['Category'] = bios.apply(categorize_athlete, axis=1)
```
